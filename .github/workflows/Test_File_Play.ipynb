One of the last things we can look at before modeling, is the correlation between variables to address potential overfitting. This can be done below by hitting explore data frame and correlation matrix. Also make sure you convert Donator ID to a string.

```python
y['Donate ID'] = y['Donate ID'].astype('string')
y['Ice Cream Products Consumed Per Week'] = y['Ice Cream Products Consumed Per Week'].astype('Int64')
y[['Ice Cream Products Consumed Per Week']] = y[['Ice Cream Products Consumed Per Week']].fillna(y[['Ice Cream Products Consumed Per Week']].median())
y
```

Lets try to run a quick KNN Model to see if we can cluster our groups by their device color preference. The first part we will separate the data set into a train and test (so we build the algorithm on one and test the accuracy on the other). In order to to that we need to split our dependent variable from our indepedent variable. We will make x all the independent variables and z the dependent variables. (for this model we will only use numerical variables as we should be coding string variables into dummy variables for a knn model. as you will see there is an option for one hot encoding which will make your life easier to do that.)

```python
y
x = y[['Deposit Amount', 'Ice Cream Products Consumed Per Week', 'How many desserts do you eat a week']]
z = y[['Preferred Color of Device']]
z
```

Here is the data set split code and model.You can test out varying n_neighbor values but for now I am just starting with 5

```python
from sklearn.model_selection import train_test_split
x_train, x_test, z_train, z_test = train_test_split(x, z, test_size=0.20)

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train, z_train)
```

Now we can look at our results. The first part of the code gets us predictions while second part gets us our confusion matrix

```python
z_pred = classifier.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(z_test, z_pred))
print(classification_report(z_test, z_pred))
```

We can also visualizae the results as well

Overall we can this model isn't very good as the accuracy is relatively low (~17%) and precision and recall is weak on all the colors.
